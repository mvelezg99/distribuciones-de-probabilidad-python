## 1. Prueba para un ajuste uniforme
Chris Columbus, director de mercadeo de Seven Seas, tiene la
responsabilidad de controlar el nivel de existencias para cuatro tipos de
botes vendidos por su firma. En el pasado ha ordenado nuevos botes bajo la
premisa de que los cuatro tipos son igualmente populares y la demanda de
cada tipo es la misma. Sin embargo, recientemente las existencias se han
vuelto más difíciles de controlar, y Chris considera que debería probar su
hipótesis respecto a una demanda uniforme. Sus hipótesis son:

**Ho
: la demanda es uniforme para los cuatro tipos de datos.**

**HA
: la demanda no es uniforme para los cuatro tipos de datos.**

Para probar la hipótesis, Chris selecciona una muestra de n = 48 botes
vendidos durante los últimos meses. Si la demanda es uniforme, puede
esperar que 48/4=12 botes de cada tipo se vendan. A continuación se muestra
esta expectativa junto con el número de cada tipo que en realidad de vendió.




Chris debe determinar ahora si los números vendidos realmente en cada una
de las categorías K = 4 está lo suficientemente cerca de lo que se esperaría si la
demanda fuese uniforme.

```python
n_botes = 48 # Tamaño de la muestra
k_botes = 4 # Número de categorías o clases
m_botes = 0 # Número de parámetros a estimar

oi_botes = [15, 11, 10, 12] # Frecuencia de los eventos observados
ei_botes = [12, 12, 12, 12] # Frecuencia de los eventos esperados

sign_botes = 0.05 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** La distribución poblacional es uniforme.

**Ha:** La distribución poblacional no es uniforme.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_botes = k_botes - m_botes - 1 # Grados de libertad
```

```python
crit_botes = ms.hypothesis.crit_val_chi2(df_botes, sign_botes)
crit_botes
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_botes = ms.categorical.get_x2(oi_botes, ei_botes)
chi2_botes
```

```python
distr_botes = ss.chi2(df_botes)
ms.graph.hypothesis(distr_botes, chi2_botes, sign_botes, 'right')
ms.hypothesis.reject_h0(crit_botes, chi2_botes, 'right')
```

Entonces, teniendo en cuenta que el valor del estadístico de prueba **1.1666666666666667** es **menor** al valor crítico que calculamos **7.814727903251179**. Podemos concluir con un nivel de significancia del *5%* que **no se puede rechazar la hipótesis nula.**

En conlusión, podemos decir que la distribución poblacional de las demandas es uniforme, por lo que los 4 tipos de botes son igualmente populares entre ellos.

## 2. Prueba de ajuste de un parámetro específico
El John Dillinger National Bank, en New York, trata de seguir una política
de extender un 60% sus créditos a empresas comerciales, un 10% a
personas naturales y un 30% a prestatarios extranjeros.

Para determinar si la política se está siguiendo, Jay Hoover, vicepresidente
de mercadeo, selecciona aleatoriamente 85 créditos que se aprobaron
recientemente. Encuentra que 62 de tales créditos se otorgaron a
negocios, 10 a personas naturales, y 13 a prestatarios extranjeros. Al nivel
del 10%, ¿parece que el patrón de cartera deseado de preserva?

**Ho : se mantuvo el patrón deseado: 60% son créditos comerciales, 10% son préstamos personales y 30% son créditos extranjeros.**

**HA : el patrón deseado no se mantuvo.**



```python
n_creditos = 85 # Tamaño de la muestra
k_creditos = 3 # Número de categorías o clases
m_creditos = 0 # Número de parámetros a estimar

pi_creditos = [0.6, 0.10, 0.30] # Frecuencia de probabilidades esperadas

oi_creditos = [62, 10, 13] # Frecuencia de los eventos observados
ei_creditos = ms.categorical.get_ei(oi_creditos, pi_creditos) # Frecuencia de los eventos esperados

sign_creditos = 0.1 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** Se mantuvo el patrón deseado.

**Ha:** No se mantuvo el patrón deseado.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_creditos = k_creditos - m_creditos - 1 # Grados de libertad
```

```python
crit_creditos = ms.hypothesis.crit_val_chi2(df_creditos, sign_creditos)
crit_creditos
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_creditos = ms.categorical.get_x2(oi_creditos, ei_creditos)
chi2_creditos
```

```python
distr_creditos = ss.chi2(df_creditos)
ms.graph.hypothesis(distr_creditos, chi2_creditos, sign_creditos, 'right')
ms.hypothesis.reject_h0(crit_creditos, chi2_creditos, 'right')
```

**SI debemos rechazar la hipótesis nula**. Teniendo en cuenta que el estadístico de prueba **8.764705882352942** es **mayor** al valor crítico que obtuvimos **4.605170185988092**, entonces cae en la zona de rechazo con un nivel de significancia del *10%*.

Podemos concluir entonces, que el patrón que deseaba el banco no se mantiene, por lo que no se está siguiendo la política que proponen.
