## 1. Prueba para un ajuste uniforme
Chris Columbus, director de mercadeo de Seven Seas, tiene la
responsabilidad de controlar el nivel de existencias para cuatro tipos de
botes vendidos por su firma. En el pasado ha ordenado nuevos botes bajo la
premisa de que los cuatro tipos son igualmente populares y la demanda de
cada tipo es la misma. Sin embargo, recientemente las existencias se han
vuelto más difíciles de controlar, y Chris considera que debería probar su
hipótesis respecto a una demanda uniforme. Sus hipótesis son:

***Ho
: la demanda es uniforme para los cuatro tipos de datos.***

***HA
: la demanda no es uniforme para los cuatro tipos de datos.***

Para probar la hipótesis, Chris selecciona una muestra de n = 48 botes
vendidos durante los últimos meses. Si la demanda es uniforme, puede
esperar que 48/4=12 botes de cada tipo se vendan. A continuación se muestra
esta expectativa junto con el número de cada tipo que en realidad de vendió.




Chris debe determinar ahora si los números vendidos realmente en cada una
de las categorías K = 4 está lo suficientemente cerca de lo que se esperaría si la
demanda fuese uniforme.

```python
n_botes = 48 # Tamaño de la muestra
k_botes = 4 # Número de categorías o clases
m_botes = 0 # Número de parámetros a estimar

oi_botes = [15, 11, 10, 12] # Frecuencia de los eventos observados
ei_botes = [12, 12, 12, 12] # Frecuencia de los eventos esperados

sign_botes = 0.05 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** La distribución poblacional es uniforme.

**Ha:** La distribución poblacional no es uniforme.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_botes = k_botes - m_botes - 1 # Grados de libertad
```

```python
crit_botes = ms.hypothesis.crit_val_chi2(df_botes, sign_botes)
crit_botes
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_botes = ms.categorical.get_x2(oi_botes, ei_botes)
chi2_botes
```

```python
distr_botes = ss.chi2(df_botes)
ms.graph.hypothesis(distr_botes, chi2_botes, sign_botes, 'right')
ms.hypothesis.reject_h0(crit_botes, chi2_botes, 'right')
```

Entonces, teniendo en cuenta que el valor del estadístico de prueba **1.1666666666666667** es **menor** al valor crítico que calculamos **7.814727903251179**. Podemos concluir con un nivel de significancia del *5%* que **no se puede rechazar la hipótesis nula.**

En conlusión, podemos decir que la distribución poblacional de las demandas es uniforme, por lo que los 4 tipos de botes son igualmente populares entre ellos.

## 2. Prueba de ajuste de un parámetro específico
El John Dillinger National Bank, en New York, trata de seguir una política
de extender un 60% sus créditos a empresas comerciales, un 10% a
personas naturales y un 30% a prestatarios extranjeros.

Para determinar si la política se está siguiendo, Jay Hoover, vicepresidente
de mercadeo, selecciona aleatoriamente 85 créditos que se aprobaron
recientemente. Encuentra que 62 de tales créditos se otorgaron a
negocios, 10 a personas naturales, y 13 a prestatarios extranjeros. Al nivel
del 10%, ¿parece que el patrón de cartera deseado de preserva?

***Ho : se mantuvo el patrón deseado: 60% son créditos comerciales, 10% son préstamos personales y 30% son créditos extranjeros.***

***HA : el patrón deseado no se mantuvo.***



```python
n_creditos = 85 # Tamaño de la muestra
k_creditos = 3 # Número de categorías o clases
m_creditos = 0 # Número de parámetros a estimar

pi_creditos = [0.60, 0.10, 0.30] # Frecuencia de probabilidades esperadas

oi_creditos = [62, 10, 13] # Frecuencia de los eventos observados
ei_creditos = ms.categorical.get_ei(oi_creditos, pi_creditos) # Frecuencia de los eventos esperados

sign_creditos = 0.1 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** Se mantuvo el patrón deseado.

**Ha:** No se mantuvo el patrón deseado.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_creditos = k_creditos - m_creditos - 1 # Grados de libertad
```

```python
crit_creditos = ms.hypothesis.crit_val_chi2(df_creditos, sign_creditos)
crit_creditos
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_creditos = ms.categorical.get_x2(oi_creditos, ei_creditos)
chi2_creditos
```

```python
distr_creditos = ss.chi2(df_creditos)
ms.graph.hypothesis(distr_creditos, chi2_creditos, sign_creditos, 'right')
ms.hypothesis.reject_h0(crit_creditos, chi2_creditos, 'right')
```

**SI debemos rechazar la hipótesis nula**. Teniendo en cuenta que el estadístico de prueba **8.764705882352942** es **mayor** al valor crítico que obtuvimos **4.605170185988092**, entonces cae en la zona de rechazo con un nivel de significancia del *10%*.

Podemos concluir entonces, que el patrón que deseaba el banco no se mantiene, por lo que no se está siguiendo la política que proponen.

## 3. Prueba de normalidad
Las especificaciones para la producción de tanques de aire utilizados en
inmersión requieren que los tanques se llenen a una presión promedio de
600 psi. Se permite una desviación estándar de 10 psi. Las especificaciones de
seguridad permiten una distribución normal en los niveles de llenado.
Usted acaba de ser contratado por Aqua Lung, su primera tarea es determinar
si los niveles de llenado se ajustan a una distribución normal. Aqua Lung está
seguro de que la media de 600 psi y la desviación estándar de 10 psi
prevalecen. Solo queda probar la naturaleza de la distribución. Para esto se
mide n = 1000 tanques y se halla la siguiente distribución:



Las hipótesis son:

**Ho: los niveles de llenado están distribuidos normalmente.**

**HA: los niveles de llenado no están distribuidos normalmente.**

Se desea probar la hipótesis al nivel del 5%.

```python
mu_aqua = 600 # Media poblacional
sigma_aqua = 10 # Desviación estándar poblacional

n_aqua = 1000 # Tamaño de la muestra
k_aqua = 6 # Número de categorías o clases
m_aqua = 0 # Número de parámetros a estimar

intervals_aqua = [
    (0, 580),
    (580, 590),
    (590, 600),
    (600, 610),
    (610, 620),
    (620, float('inf'))
]

oi_aqua = [20, 142, 310, 370, 128, 30] # Frecuencia de los eventos observados
pi_aqua = ms.categorical.get_pi(ss.norm(mu_aqua, sigma_aqua), intervals_aqua, oi_aqua) # Frecuencia de probabilidades esperadas
ei_aqua = ms.categorical.get_ei(oi_aqua, pi_aqua) # Frecuencia de los eventos esperados

sign_aqua = 0.05 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** Se mantiene una distribución normal.

**Ha:** No se mantiene una distribución normal.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_aqua = k_aqua - m_aqua - 1
```

```python
crit_aqua = ms.hypothesis.crit_val_chi2(df_aqua, sign_aqua)
crit_aqua
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_aqua = ms.categorical.get_x2(oi_aqua, ei_aqua)
chi2_aqua
```

```python
distr_aqua = ss.chi2(df_aqua)
ms.graph.hypothesis(distr_aqua, chi2_aqua, sign_aqua, 'right')
ms.hypothesis.reject_h0(crit_aqua, chi2_aqua, 'right')
```

**NO se rechaza la hipótesis nula**. Dado que el estadístico de prueba **8.659793636460499** es **menor** al valor crítico calculado **11.070497693516351**, por lo que cae en la zona de no rechazo con un nivel de significancia del *5%*.

Concluimos entonces que los niveles de llenado que tienen en Aqua Lung siguen en efecto una distribución normal.

## 4. Tablas de contingencia. Una prueba de independencia
Wilma Keeto es la directora de investigación de productos en Dow
Chemical. En su proyecto actual, la señorita Keeto debe determinar si existe
alguna relación entre la clasificación de efectividad que los consumidores
asignan a un nuevo insecticida y el sitio (urbano o rural) en el cual se utiliza.
De los 100 consumidores a quienes se les practicó la encuesta, 75 vivían en
zonas urbanas y 25 en zonas rurales.



La señorita Keeto desea comparar el atributo B (ubicación) con el atributo A
(clasificación del producto). Sus hipótesis son:

***Ho : la clasificación y la ubicación son independientes.***

***HA : la clasificación y la ubicación no son independientes.***

Si la ubicación no tiene ningún impacto en la clasificación efectiva, entonces el
porcentaje de residentes urbanos que clasificaron el producto “por encima del
promedio” debería ser igual al porcentaje de residentes rurales que clasificaron
el producto “por encima del promedio”.

Wilma fija un nivel de significancia del 10%.

```python
n_keeto = 100 # Tamaño de la muestra
r_keeto = 3 # Número de filas
c_keeto = 2 # Número de columnas

oi_keeto = [[20, 40, 15], [11, 8, 6]] # Frecuencia de los eventos observados
ei_keeto = ms.categorical.contingency_ei(oi_keeto) # Frecuencia de los eventos esperados

sign_keeto = 0.10 # Nivel de significancia
```

* **Plantear hipótesis**

**Ho:** Los atributos son independientes.

**Ha:** Los atributos no son independientes.

- **Ahora encontraremos el valor crítico para rechazar la hipótesis nula**

```python
df_keeto = (r_keeto - 1) * (c_keeto - 1)
```

```python
crit_keeto = ms.hypothesis.crit_val_chi2(df_keeto, sign_keeto)
crit_keeto
```

- **Ahora encontraremos el estadístico de prueba (CHI CUADRADO) para comparar con el valor crítico y determinar si se rechaza la hipótesis nula**

```python
chi2_keeto = ms.categorical.get_x2(oi_keeto, ei_keeto)
chi2_keeto
```

```python
distr_keeto = ss.chi2(df_keeto)
ms.graph.hypothesis(distr_keeto, chi2_keeto, sign_keeto, "right")
ms.hypothesis.reject_h0(crit_keeto, chi2_keeto, "right")
```

**NO se rechaza la hipótesis nula**. Teniendo en cuenta que el estadístico de prueba **3.7378392217101895**, es menor al valor crítico calculado **4.605170185988092**, por lo que cae en la zona de no rechazo con un nivel de significancia del *10%*.

Por lo que se puede concluir que la ubicación de los consumidores no tiene por qué afectar en la calificación que le dieron al producto.

```python
crit_auto = ms.hypothesis.crit_val_chi2(df_auto, sign_auto)
crit_auto
```
