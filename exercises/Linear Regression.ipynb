## 1. Regresión lineal simple
La agencia de Hop Scotch Airlines , la aerolínea transportadora más pequeña del mundo,
considera que existe una relación directa entre los gastos publicitarios y el número de
pasajeros que escogen viajar por Hop Scotch. Para determinar si esta relación existe, y si es
así cuál podría ser la naturaleza exacta, los estadísticos empleados por Hop Scotch
decidieron utilizar los procedimientos MCO para determinar el modelo de regresión.
Se recolectaron los valores mensuales por gastos de publicidad y número de pasajeros
para los n = 15 meses más recientes.



```python
x_hop = [10, 12, 8, 17, 10, 15, 10, 14, 19, 10, 11, 13, 16, 10, 12] # Variable independiente
y_hop = [15, 17, 13, 23, 16, 21, 14, 20, 24, 17, 16, 18, 23, 15, 16] # Variable dependiente
```

```python
ms.graph.linregr(x_hop, y_hop)
```

**Si se invierte US$10000 en publicidad; ¿cuánto dinero de los pasajeros se esperaría obtener?**

```python
model_hop = ms.linregr.regression(x_hop, y_hop)
model_hop(10)
```

Se esperaría obtener US$15200 en dinero proveniente de los pasajeros gracias a la publicidad invertida.

El coeficiente de 1.08 significa que por cada incremento de una unidad en
X, Y aumentará en 1.08 unidades.

**¿Cuál sería el error estándar para el caso de Hop Scotch Airlines**

```python
ms.linregr.get_se(x_hop, y_hop)
```

El error estándar de Hop Scotch Airlines que tienen en su modelo de regresión lineal es de 0.907 miles de pasajeros.

## 2. Regresión lineal simple
Un economista del departamento de recursos humanos de Florida State
está preparando un estudio sobre el comportamiento del consumidor. Él recolectó los
datos que aparecen en miles de dólares para determinar si existe una relación entre el
ingreso del consumidor y los niveles de consumo.

c. Calcule e interprete el modelo de regresión. ¿Qué le dice este modelo sobre la relación

invierte en consumo?

## 4. Pruebas de hipótesis para la pendiente poblacional
Para el caso de Hop Scotch, ¿Se puede realizar una regresión lineal?

```python
sign_hop = 0.05 # Nivel de significancia
```

* **Plantear las hipótesis**

$H_0: β_1 = 0$

$H_A: β_1 ≠ 0$

* **Valores críticos**

```python
df_hop = len(x_hop) - 2
```

```python
crit_hop = ms.hypothesis.crit_val_t(df_hop, sign_hop, 'two')
crit_hop
```

* **Estadístico de prueba (T)**

```python
t_hop = ms.linregr.get_t_beta(x_hop, y_hop)
t_hop
```

```python
distr_hop = ss.t(df_hop)
ms.graph.hypothesis(distr_hop, t_hop, sign_hop, 'two')
ms.hypothesis.reject_h0(crit_hop, t_hop, 'two')
```

**SI se rechaza la hipótesis nula.** Teniendo en cuenta que el estadístico de prueba *13.994917613628866* es mayor al valor crítico calculado *2.1603686564610127*. Por lo que cae en la zona de rechazo con un nivel de significancia del *5%*.

Por lo que podemos concluir que parece existir una relación entre las variables, 

```python
ms.linregr.get_r(x_hop, y_hop)
```

```python
ms.linregr.get_r2(x_hop, y_hop)
```

```python
ms.intervals.beta(x_hop, y_hop, 0.95)
```
